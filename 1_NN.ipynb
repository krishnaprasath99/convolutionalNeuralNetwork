## Part-1(Building a basic NN)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

### Reading the dataset

```python
df=pd.read_csv('dataset.csv')
print(df)
```

### Dropping the samples with categorical values in the numerical features and changing the numerical datatype to float

```python
alpha='abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
alpha=list(alpha)
df=df[~df.isin(alpha)]
df=df.dropna()
df['f1']=df['f1'].astype('float')
df['f2']=df['f2'].astype('float')
df['f4']=df['f4'].astype('float')
df['f5']=df['f5'].astype('float')
df['f6']=df['f6'].astype('float')
df['f7']=df['f7'].astype('float')
df.describe()
```

### Scaling the numerical features to have zero mean and unit variance.

```python
target_df = df.target
print(target_df)
df = df.drop('target',axis='columns')
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
df = pd.DataFrame(scaled_data, columns = ['f1','f2','f3','f4','f5','f6','f7'])
df = pd.concat([df, target_df], axis=1)
print(df)
```

### Boxplot to check the outliers present in the dataset

```python
plt.figure(figsize=(14, 10))
sns.boxplot(data=df)
plt.xlabel("Values")
plt.title("Box Plot to check outliers")
plt.show()
```

### Correlation plot to know the correlation between the features

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

### Removing the outliers present in the dataset using Inter-quartile Range

```python
#REMOVE OUTLIERS
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

threshold = 1
outliers = ((df < (Q1 - threshold * IQR)) | (df > (Q3 + threshold * IQR))).any(axis=1)
cleaned_data = df[~outliers]
df = cleaned_data

print(cleaned_data)
plt.figure(figsize=(14, 10))
sns.boxplot(data=cleaned_data)
plt.xlabel("Values")
plt.title("Box Plot after removing potential outliers")
plt.show()
```

### Splitting the dataset into train,test and validation

```python
from sklearn.model_selection import train_test_split 
X = df.drop('target',axis='columns')
y = df.target
X_train, X_test, y_train, y_test = train_test_split(X,y , random_state=104, test_size=0.10, shuffle=True)
X_train, X_val, y_train, y_val = train_test_split(X_train,y_train , random_state=104, test_size=0.10, shuffle=True)
```

### Defining the neural network architecture

```python
import torch
import torch.nn as nn
import torch.functional as F

class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork,self).__init__()
        self.neuralNet=nn.Sequential(
            nn.Linear(7,128),
            nn.ReLU(),
            nn.Linear(128,64),
            nn.ReLU(),
            nn.Dropout(p=0.3),
            nn.Linear(64,1),
            nn.Sigmoid()
        )
    def forward(self,x):
        output=self.neuralNet(x)
        return output
    
model=NeuralNetwork()
print(model)
```

### Printing the summary of the model

```python
from torchinfo import summary

summary(model)
```

### Converting the training,validation and testing datasets to tensor form

```python
X_train_tensor=torch.tensor(X_train.to_numpy(),dtype=torch.float32)
X_val_tensor=torch.tensor(X_val.to_numpy(),dtype=torch.float32)
y_train_tensor=torch.tensor(y_train.to_numpy(),dtype=torch.float32)
y_val_tensor=torch.tensor(y_val.to_numpy(),dtype=torch.float32)
X_test_tensor=torch.tensor(X_test.to_numpy(),dtype=torch.float32)
y_test_tensor=torch.tensor(y_test.to_numpy(),dtype=torch.float32)
```

### Defining the batch size

```python
batch_size=102
```

```python
for param in model.parameters():
    print(param.requires_grad)
```

### Function to calculate the model accuracy on the dataset X and y that is passed as parameters to the function

```python
from torchmetrics.classification import BinaryAccuracy
def acc(model,X,y):
    metric=BinaryAccuracy() 
    y_pred=model(X)
    predicted_y=y_pred.round().flatten()
    return (metric(predicted_y,y)*100)
```

### Function to train the model and calculate the losses and accuracy for each epoch

```python
from torch.optim import Adam
import time
def train(model):
        train_loss=[]
        vald_loss=[]
        testi_loss=[]
        train_acc=[]
        val_acc=[]
        test_acc=[]
        criterion=nn.BCELoss()
        optimizer=Adam(model.parameters(),lr=10**(-4))
        start_time=time.time()
        for epoch in range(1000):
                for i in range(0,len(X_train_tensor),batch_size):
                        X_train_batch=X_train_tensor[i:i+batch_size]
                        y_train_pred=model(X_train_batch).flatten()
                        optimizer.zero_grad()
                        y_train_batch=y_train_tensor[i:i+batch_size]
                        loss=criterion(y_train_pred,y_train_batch)
                        loss.backward()
                        optimizer.step()
                train_loss.append(loss.item())
                
                #print("Training loss obtained is:{0}".format(loss))
                model.eval()
                val_loss=0
                with torch.no_grad():
                        for j in range(0,len(X_val_tensor),batch_size):
                                X_val_batch=X_val_tensor[j:j+batch_size]
                                y_val_pred=model(X_val_batch).flatten()
                                optimizer.zero_grad()
                                y_val_batch=y_val_tensor[j:j+batch_size]
                                val_loss=criterion(y_val_pred,y_val_batch)   
                        vald_loss.append(val_loss.item())      

                #print("Validation loss obtained is{0}".format(val_loss))
                        for j in range(0,len(X_test_tensor),batch_size):
                                X_test_batch=X_test_tensor[j:j+batch_size]
                                y_test_pred=model(X_test_batch).flatten()
                                optimizer.zero_grad()
                                y_test_batch=y_test_tensor[j:j+batch_size]
                                test_loss=criterion(y_test_pred,y_test_batch)   
                        testi_loss.append(test_loss.item())      
                #print("Testing loss obtained is{0}".format(test_loss))

                #Accuracy
                        train_acc.append(acc(model,X_train_tensor,y_train_tensor))
                        val_acc.append(acc(model,X_val_tensor,y_val_tensor))
                        test_acc.append(acc(model,X_test_tensor,y_test_tensor))
        end_time=time.time()
        return start_time,end_time,train_loss,vald_loss,testi_loss,train_acc,val_acc,test_acc

start_time,end_time,train_loss,vald_loss,test_loss,train_acc,val_acc,test_acc=train(model=model)
```

## Part-2(Optimizing Neural Network)

### Defining 9 different models using base model in part-1 with one hyperparameter changed for each model 

```python
#Part-2
import torch.nn as nn
import torch.nn.init as init

def init_weight1(layer):
    if isinstance(layer, nn.Linear):
        init.uniform_(layer.weight, a=-0.2, b=0.2)  
        layer.bias.data.fill_(0)

def init_weight2(layer):
    if isinstance(layer, nn.Linear):
        init.kaiming_normal_(layer.weight, mode='fan_in', nonlinearity='relu')
        layer.bias.data.fill_(0)

def init_weight3(layer):
    if isinstance(layer, nn.Linear):
        init.xavier_normal_(layer.weight)
        layer.bias.data.fill_(0)


class NeuralNetworkDropout1(nn.Module):
    def __init__(self):
        super(NeuralNetworkDropout1,self).__init__()
        self.neuralNet=nn.Sequential(
            nn.Linear(7,128),
            nn.ReLU(),
            nn.Linear(128,64),
            nn.ReLU(),
            nn.Dropout(p=0.45),
            nn.Linear(64,1),
            nn.Sigmoid()
        )
    def forward(self,x):
        output=self.neuralNet(x)
        return output

class NeuralNetworkDropout2(nn.Module):
    def __init__(self):
        super(NeuralNetworkDropout2,self).__init__()
        self.neuralNet=nn.Sequential(
            nn.Linear(7,128),
            nn.ReLU(),
            nn.Linear(128,64),
            nn.ReLU(),
            nn.Dropout(p=0.5),
            nn.Linear(64,1),
            nn.Sigmoid()
        )
    def forward(self,x):
        output=self.neuralNet(x)
        return output

class NeuralNetworkDropout3(nn.Module):
    def __init__(self):
        super(NeuralNetworkDropout3,self).__init__()
        self.neuralNet=nn.Sequential(
            nn.Linear(7,128),
            nn.ReLU(),
            nn.Linear(128,64),
            nn.ReLU(),
            nn.Dropout(p=0.55),
            nn.Linear(64,1),
            nn.Sigmoid()
        )
    def forward(self,x):
        output=self.neuralNet(x)
        return output

class NeuralNetworkActivationFn1(nn.Module):
    def __init__(self):
        super(NeuralNetworkActivationFn1,self).__init__()
        self.neuralNet=nn.Sequential(
            nn.Linear(7,128),
            nn.Sigmoid(),
            nn.Linear(128,64),
            nn.Sigmoid(),
            nn.Dropout(p=0.3),
            nn.Linear(64,1),
            nn.Sigmoid()
        )
    def forward(self,x):
        output=self.neuralNet(x)
        return output

class NeuralNetworkActivationFn2(nn.Module):
    def __init__(self):
        super(NeuralNetworkActivationFn2,self).__init__()
        self.neuralNet=nn.Sequential(
            nn.Linear(7,128),
            nn.ReLU(),
            nn.Linear(128,64),
            nn.ReLU(),
            nn.Dropout(p=0.3),
            nn.Linear(64,1),
            nn.Sigmoid()
        )
    def forward(self,x):
        output=self.neuralNet(x)
        return output
    

class NeuralNetworkActivationFn3(nn.Module):
    def __init__(self):
        super(NeuralNetworkActivationFn3,self).__init__()
        self.neuralNet=nn.Sequential(
            nn.Linear(7,128),
            nn.LeakyReLU(negative_slope=0.01),
            nn.Linear(128,64),
            nn.LeakyReLU(negative_slope=0.01),
            nn.Dropout(p=0.3),
            nn.Linear(64,1),
            nn.Sigmoid()
        )
    def forward(self,x):
        output=self.neuralNet(x)
        return output 

wt_changed_1 = NeuralNetwork()
wt_changed_1.apply(init_weight1)

wt_changed_2 = NeuralNetwork()
wt_changed_2.apply(init_weight2)

wt_changed_3 = NeuralNetwork()
wt_changed_3.apply(init_weight3)

dropout_changed_1 = NeuralNetworkDropout1()
dropout_changed_1.apply(init_weight1)

dropout_changed_2 = NeuralNetworkDropout2()
dropout_changed_2.apply(init_weight1)

dropout_changed_3 = NeuralNetworkDropout3()
dropout_changed_3.apply(init_weight1)

act_fn_changed_1 = NeuralNetworkActivationFn1()
act_fn_changed_1.apply(init_weight1)

act_fn_changed_2 = NeuralNetworkActivationFn2()
act_fn_changed_2.apply(init_weight1)

act_fn_changed_3 = NeuralNetworkActivationFn3()
act_fn_changed_3.apply(init_weight1)

```

```python
start_time_1,end_time_1,train_loss_1,vald_loss_1,testi_loss_1,train_acc_1,val_acc_1,test_acc_1=train(model=wt_changed_1)
print("Accuracy obtained on test set for wt_changed_1:{0}%".format(acc(wt_changed_1,X_test_tensor,y_test_tensor).item()))

start_time_2,end_time_2,train_loss_2,vald_loss_2,testi_loss_2,train_acc_2,val_acc_2,test_acc_2=train(model=wt_changed_2)
print("Accuracy obtained on test set for wt_changed_2:{0}%".format(acc(wt_changed_2,X_test_tensor,y_test_tensor).item()))

start_time_3,end_time_3,train_loss_3,vald_loss_3,testi_loss_3,train_acc_3,val_acc_3,test_acc_3=train(model=wt_changed_3)
print("Accuracy obtained on test set for wt_changed_3:{0}%".format(acc(wt_changed_3,X_test_tensor,y_test_tensor).item()))

start_time_4,end_time_4,train_loss_4,vald_loss_4,testi_loss_4,train_acc_4,val_acc_4,test_acc_4=train(model=dropout_changed_1)
print("Accuracy obtained on test set for dropout_changed_1:{0}%".format(acc(dropout_changed_1,X_test_tensor,y_test_tensor).item()))

start_time_5,end_time_5,train_loss_5,vald_loss_5,testi_loss_5,train_acc_5,val_acc_5,test_acc_5=train(model=dropout_changed_2)
print("Accuracy obtained on test set for dropout_changed_2:{0}%".format(acc(dropout_changed_2,X_test_tensor,y_test_tensor).item()))

start_time_6,end_time_6,train_loss_6,vald_loss_6,testi_loss_6,train_acc_6,val_acc_6,test_acc_6=train(model=dropout_changed_3)
print("Accuracy obtained on test set for dropout_changed_3:{0}%".format(acc(dropout_changed_3,X_test_tensor,y_test_tensor).item()))

start_time_7,end_time_7,train_loss_7,vald_loss_7,testi_loss_7,train_acc_7,val_acc_7,test_acc_7=train(model=act_fn_changed_1)
print("Accuracy obtained on test set for act_fn_changed_1:{0}%".format(acc(act_fn_changed_1,X_test_tensor,y_test_tensor).item()))

start_time_8,end_time_8,train_loss_8,vald_loss_8,testi_loss_8,train_acc_8,val_acc_8,test_acc_8=train(model=act_fn_changed_2)
print("Accuracy obtained on test set for act_fn_changed_2:{0}%".format(acc(act_fn_changed_2,X_test_tensor,y_test_tensor).item()))

start_time_9,end_time_9,train_loss_9,vald_loss_9,testi_loss_9,train_acc_9,val_acc_9,test_acc_9=train(model=act_fn_changed_3)
print("Accuracy obtained on test set for act_fn_changed_3:{0}%".format(acc(act_fn_changed_3,X_test_tensor,y_test_tensor).item()))
```

### Choosing the best accuracy model among the 10 models(base+ 9 setups) that was obtained above and trying different methods to improve accuracy

### 1.Learning rate scheduling

```python
#Training the model with Learning rate scheduler
from torch.optim.lr_scheduler import ExponentialLR
def train_with_lrs(model):
        train_loss=[]
        vald_loss=[]
        testi_loss=[]
        train_acc=[]
        val_acc=[]
        test_acc=[]
        criterion=nn.BCELoss()
        optimizer=Adam(model.parameters(),lr=10**(-2))
        scheduler=ExponentialLR(optimizer, gamma=0.9)
        start_time=time.time()
        for epoch in range(1000):
                for i in range(0,len(X_train_tensor),batch_size):
                        X_train_batch=X_train_tensor[i:i+batch_size]
                        y_train_pred=model(X_train_batch).flatten()
                        optimizer.zero_grad()
                        y_train_batch=y_train_tensor[i:i+batch_size]
                        loss=criterion(y_train_pred,y_train_batch)
                        loss.backward()
                        optimizer.step()
                train_loss.append(loss.item())
                
                #print("Training loss obtained is:{0}".format(loss))
                model.eval()
                with torch.no_grad():
                        for j in range(0,len(X_val_tensor),batch_size):
                                X_val_batch=X_val_tensor[j:j+batch_size]
                                y_val_pred=model(X_val_batch).flatten()
                                optimizer.zero_grad()
                                y_val_batch=y_val_tensor[j:j+batch_size]
                                val_loss=criterion(y_val_pred,y_val_batch)
                        vald_loss.append(val_loss.item())      

                #print("Validation loss obtained is{0}".format(val_loss))
                        for j in range(0,len(X_test_tensor),batch_size):
                                X_test_batch=X_test_tensor[j:j+batch_size]
                                y_test_pred=model(X_test_batch).flatten()
                                optimizer.zero_grad()
                                y_test_batch=y_test_tensor[j:j+batch_size]
                                test_loss=criterion(y_test_pred,y_test_batch)   
                        testi_loss.append(test_loss.item())      
                #print("Testing loss obtained is{0}".format(test_loss))
                #Accuracy
                        train_acc.append(acc(model,X_train_tensor,y_train_tensor))
                        val_acc.append(acc(model,X_val_tensor,y_val_tensor))
                        test_acc.append(acc(model,X_test_tensor,y_test_tensor))
                scheduler.step()
        end_time=time.time()
        return start_time,end_time,train_loss,vald_loss,testi_loss,train_acc,val_acc,test_acc


```

### 4.Early stopping technique

```python
import numpy as np
class EarlyStopping:
    def __init__(self, patience=5, min_delta=0):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = np.Inf
        self.early_stop = False

    def __call__(self, current_loss):
        if current_loss - self.best_loss > self.min_delta:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_loss = current_loss
            self.counter = 0

def trainEarlyStopping(model, X_train_tensor, y_train_tensor, X_val_tensor, y_val_tensor, X_test_tensor, y_test_tensor, batch_size=32):
    train_loss = []
    vald_loss = []
    testi_loss = []
    train_acc = []
    val_acc = []
    test_acc = []
    criterion = nn.BCELoss()
    optimizer = Adam(model.parameters(), lr=1e-4)
    start_time = time.time()

    #Adding an instance to call early stopping in existing training loop and passing the parameters
    early_stopping = EarlyStopping(patience=5, min_delta=0)

    for epoch in range(1000):
        for i in range(0, len(X_train_tensor), batch_size):
            X_train_batch = X_train_tensor[i:i + batch_size]
            y_train_pred = model(X_train_batch).flatten()
            optimizer.zero_grad()
            y_train_batch = y_train_tensor[i:i + batch_size]
            loss = criterion(y_train_pred, y_train_batch)
            loss.backward()
            optimizer.step()

        train_loss.append(loss.item())

        model.eval()
        val_loss = 0
        with torch.no_grad():
            for j in range(0, len(X_val_tensor), batch_size):
                X_val_batch = X_val_tensor[j:j + batch_size]
                y_val_pred = model(X_val_batch).flatten()
                optimizer.zero_grad()
                y_val_batch = y_val_tensor[j:j + batch_size]
                val_loss += criterion(y_val_pred, y_val_batch).item()
        val_loss /= len(X_val_tensor) / batch_size
        vald_loss.append(val_loss)

        for j in range(0, len(X_test_tensor), batch_size):
            X_test_batch = X_test_tensor[j:j + batch_size]
            y_test_pred = model(X_test_batch).flatten()
            optimizer.zero_grad()
            y_test_batch = y_test_tensor[j:j + batch_size]
            test_loss = criterion(y_test_pred, y_test_batch).item()
        test_loss /= len(X_test_tensor) / batch_size
        testi_loss.append(test_loss)

        train_acc.append(acc(model, X_train_tensor, y_train_tensor))
        val_acc.append(acc(model, X_val_tensor, y_val_tensor))
        test_acc.append(acc(model, X_test_tensor, y_test_tensor))

        early_stopping(val_loss)

        if early_stopping.early_stop:
            print("Early stopping hit. Stopping training in epoch : ", epoch)
            break

    end_time = time.time()
    return start_time, end_time, train_loss, vald_loss, testi_loss, train_acc, val_acc, test_acc
```
